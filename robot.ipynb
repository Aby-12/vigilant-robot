```python
#for each country separately (Kenya, Rwanda, Uganda, Tanzania):
#correlate Overall Governance over time with these four indicators:SECURITY & RULE OF LAW, PARTICIPATION, RIGHTS & INCLUSION,FOUNDATIONS FOR ECONOMIC OPPORTUNITY and HUMAN DEVELOPMENT
#that way, you see if governance in Kenya, for example, moves more with human development than with security, etc.
# define the indicators we care about
indicators = [
    "SECURITY & RULE OF LAW",
    "PARTICIPATION, RIGHTS & INCLUSION",
    "FOUNDATIONS FOR ECONOMIC OPPORTUNITY",
    "HUMAN DEVELOPMENT"
]

# function to compute correlations for one country
def country_corr(country):
    subset = df[df["Country"] == country]
    return subset[["OVERALL GOVERNANCE"] + indicators].corr().loc["OVERALL GOVERNANCE", indicators]

# run for the four countries
for country in ["Kenya", "Rwanda", "Uganda", "Tanzania"]:
    print(f"\n{country} correlations with OVERALL GOVERNANCE")
    print(country_corr(country))
#Kenya strong correlations with Foundations for Economic Opportunity (green) and Human Development (red).weaker/negative link with Security & Rule of Law (blue).
#Rwanda governance aligns strongly with Security & Rule of Law and Human Development.Participation, Rights & Inclusion shows little or even slightly negative correlation.
#Uganda governance strongly tied to Security & Rule of Law and Participation.weaker/negative relationship with Foundations and Human Development.
#Tanzania governance strongly tied to Foundations and Human Development.weaker or even negative relationship with Participation.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# --- config: indicator columns (exact names from your df) ---
indicators = [
    "SECURITY & RULE OF LAW",
    "PARTICIPATION, RIGHTS & INCLUSION",
    "FOUNDATIONS FOR ECONOMIC OPPORTUNITY",
    "HUMAN DEVELOPMENT",
]
OG = "OVERALL GOVERNANCE"
COUNTRY = "Country"
YEAR = "Year"

# --- helper: compute correlations for one country safely ---
def country_corr(df, country, og_col=OG, inds=indicators):
    sub = df[df[COUNTRY] == country].copy()
    if sub.empty:
        raise ValueError(f"No rows found for country: {country}")
    # ensure numeric (coerce strings/objects to numbers if needed)
    for c in [og_col] + inds:
        sub[c] = pd.to_numeric(sub[c], errors="coerce")
    # need at least 2 non-NaN rows per column for corr
    corr = sub[[og_col] + inds].corr().loc[og_col, inds]
    return corr

# --- build results ---
results = {}
for c in ["Kenya", "Rwanda", "Uganda", "Tanzania"]:
    results[c] = country_corr(df, c)

corr_df = pd.DataFrame(results).T  # rows=countries, columns=indicators

# --- plot ---
ax = corr_df.plot(kind="bar", figsize=(12, 7))
ax.set_title("Correlation of OVERALL GOVERNANCE with key indicators")
ax.set_ylabel("Correlation coefficient")
ax.set_xlabel("Country")
ax.tick_params(axis="x", rotation=0)
ax.legend(title="Indicator")
plt.tight_layout()

# --- save (make sure folder exists) ---
Path("figures").mkdir(exist_ok=True)
ax.figure.savefig("figures/country_correlations.png", dpi=300)

plt.show()
```

```python
#3.heatmaps for correlations between indicators
#this is a matrix view of how strongly the indicators are related.
import seaborn as sns

# keep only numeric columns
num_df = df.select_dtypes(include="number")

# compute correlation matrix
corr_matrix = num_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Governance Indicators")
plt.tight_layout()
plt.show()

#  save figure
plt.savefig("figures/country_Correlation of key indicators.png", dpi=300)

#what the heatmap shows
#each cell = correlation between two indicators (e.g., Overall Governance vs Education).
#red cells = strong positive correlation (they move together).
#blue cells = negative correlation (when one goes up, the other goes down).
#white/light = weak or no correlation.
#the diagonal (dark red) is always 1.0, because an indicator is perfectly correlated with itself.
```

```python
#year-on-year changes
#instead of just levels, check how scores change annually.
#e.g., does Kenya improve steadily, while Uganda declines sharply?
df["change"] = df.groupby("Country")["OVERALL GOVERNANCE"].diff()
```

```python
df[["Country","Year","OVERALL GOVERNANCE","change"]].head(12)
```

```python
import matplotlib.pyplot as plt

for country in ["Kenya","Rwanda","Uganda","Tanzania"]:
    subset = df[df["Country"]==country]
    plt.plot(subset["Year"], subset["change"], marker="o", label=country)

plt.axhline(0, color="grey", linestyle="--")  # baseline
plt.title("Year-on-Year Change in Overall Governance")
plt.xlabel("Year")
plt.ylabel("Change in score")
plt.legend()
plt.tight_layout()
plt.show()

#  save figure
plt.savefig("figures/country_year on year change.png", dpi=300)

#how to read it
#x-axis = years
#y-axis = change from the previous year (positive = improvement, negative = decline)
#dashed grey line at 0 = baseline:above  governance improved and below  governance worsene
#1. Kenya (blue line)
#lots of up-and-down movement around zero.
#generally positive changes, but some years of decline too.
#interpretation: governance is improving overall, but progress is uneven year to year.
# 2. Rwanda (orange line)
#also fluctuates, but often hovers near zero.
#some small improvements, some small declines.
#interpretation: governance is stable, with no large shifts year to year.
#3. Uganda (green line)
#consistently more time below zero.
#many years of decline, only occasional small improvements.
#interpretation: governance is on a downward trajectory, with repeated setbacks.
# 4. Tanzania (red line)
#quite volatile  sharp rises and sharp falls.
#some of the biggest positive improvements (above zero), but also some steep declines.
#interpretation: governance is highly unstable, swinging between strong gains and setbacks.
```
